```python
series1 = {'2000': 1901,
  '2001': 1999,
  '2002': 1989,
  '2003': 1351,
  '2004': 1243,
  '2005': 1135,
  '2006': 1289,
  '2007': 1387,
  '2008': 1738,
  '2009': 1528,
  '2010': 1472,
  '2011': 0,
  '2012': 400,
  '2013': 1324,
  '2014': 1504,
  '2015': 1067,
  '2016': 1347}

series2 = {'2000': 12,
  '2001': 19,
  '2002': 13,
  '2003': 5,
  '2004': 18,
  '2005': 8,
  '2006': 2,
  '2007': 5,
  '2008': 11,
  '2009': 11,
  '2010': 33,
  '2011': 37,
  '2012': 34,
  '2013': 12,
  '2014': 20,
  '2015': 15,
  '2016': 29}
```

```python
import pandas as pd

df = pd.DataFrame([series1, series2]).T
df.columns = ['newspaper', 'magazine']
df
```

```python
import scipy.stats as stats

stats.pearsonr(df.newspaper, df.magazine)
```

```python
from statsmodels.tsa.stattools import grangercausalitytests

grangercausalitytests(df, maxlag=3)
```
